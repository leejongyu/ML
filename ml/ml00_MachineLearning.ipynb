# 머신러닝이란?

* 데리터로부터 학습하도록 컴퓨터를 프로그래밍 하는 것.
* ex) 이미지 분석/ 자동 분류, 뇌종양 진단, 신문기사 자동 분류, 챗봇, 수익 예측, 음성인식, 고객 분류, 상품 추천, 게임 등....
* AI(Artificial Intelligence) **[넓은 개념]** > Machine Learning > Deep Learning **[심화된 개념]**

# Why Machine Learning
* 스팸 필터 [전통적인 프로그래밍 방법]
    * 1. 스팸에 어떤 단어(패턴)들이 주로 나타나는지 조사
    * 2. 발견한 패턴을 감지하는 알고리즘을 작성하여 이러한 패턴이 발견되면 스팸으로 분류하게 함
    * 3. 프로그램을 테스트하고, 충분한 성능(학습)이 될 때 까지 1단계와 2단계를 반복

* 전통적인 프로그래밍 방법의 한계 vs 머신 러닝
    * [한계] 규칙이 점점 길고 복잡해지므로 유지 보수가 매우 힘듬
        * -> [머신러닝] 스팸메일을 판단하는 기준을 데이터가 스스로 학습
    * [한계] 스팸 발송자가 스팸 필터에 계속 단어를 바꾸면 새로운 규칙을 추가해야함.
        * -> [머신러닝] 별도 작업 없이 새로운 데이터에서 새로운 규칙을 스스로 추가

* 데이터 마이닝(Data mining)
    * 사람이 머신러닝 알고리즘이 학습한 것을 조사하다가 예상치 못한 연관관계, 새로운 추세가 발견되기도 해서 해당 문제를 더 잘 이해하도록 도와줌.
    * 머신러닝 기술을 적용해서 대용량의 데이터를 분석하면 겉으로 보이지 않던 패턴을 발견할 수도 있다.

# 머신러닝 종류

* 사람의 감독 하에 훈련하는 것인가?
    * 지도(Supervised) : 사람이 labeling, 입력 등으로 정보, 학습하는 방법 등을 직접 알려주고 학습하도록 함
        * 분류:
        * 회귀: 예측
    * 비지도(Unsupervised) : 학습 데이터의 모든 과정에서 사람이 관여하지 않음
    * 준지도(Semi-supervised) : 학습 데이터의 일부 과정에서 사람이 관여함.
    * 강화(Reinforcement) : 

* 실시간으로 점진적인 학습을 하는가?
    * 온라인(Online) : 훈련된 결과에서 추가된 결과만 다시 학습하는 방법
    * 배치(Batch) : 누적된 데이터를 모두 학습하는 방법

* 단순하게 알고있는 데이터 포인트와 새 데이터 포인트를 비교하는 것인가, 아니면 훈련데이터 셋에서 과학자들처럼 패턴을 발견하여 예측 모델을 만드는가?
    * 사례 기반(Instance-based) 학습 : 진짜 데이터를 예시로 알려주는 것
    * 모델 기반(Model-based) 학습 : 수학적인 공식, 모델 등을 알려주는 것

# 나쁜 데이터, 나쁜 알고리즘

* 충분하지 않은 양의 훈련 데이터
* 대표성이 없는 훈련 데이터
    * sampling noise, sampling bias
* 낮은 품질의 데이터
    * 결측치(missing values, NA), 이상치(outliers)
* 관련 없는 특성
    * garbage in, garbage out
* 훈련 데이터 과대적합(Overfitting), 과소적합(Underfitting)

* Testing and Validating
    * Training set : 다양한 하이퍼 파라미터들을 갖는 여러개의 모델을 학습
    * Validation set : 최상의 성능을 내는 모델을 선택
    * Test set : 모델을 평가하고 일반화 에러를 추정
